```python
from model import FAQBot

# Initialize the bot
bot = FAQBot()

# Test the model
user_query = "ما هو مأمن؟"
results = bot.find_similar(user_query, top_k=3)

print("Results for query:", user_query)
for result in results:
    print(f"Similarity: {result['similarity']:.2f}")
    print(f"Question: {result['question']}")
    print(f"Answer: {result['answer']}")
    print("---")
```

```python
# Interactive testing
while True:
    user_input = input("\nاكتب سؤالك (أو 'exit' للخروج): ")
    if user_input.lower() == "exit":
        break
    lang = input("اختر اللغة 'ar' للعربي أو 'en' للإنجليزي (أو اضغط Enter للكشف التلقائي): ").strip().lower()
    if lang not in {"ar", "en"}:
        lang = bot.detect_language(user_input)
        print(f"تم الكشف التلقائي عن اللغة: {'العربية' if lang == 'ar' else 'الإنجليزية'}")

    results = bot.find_similar(user_input, top_k=3)
    
    if not results:
        msg = "عذرًا، لم أجد إجابة مناسبة." if lang == "ar" else "Sorry, no suitable answer found."
        print(msg)
        continue

    filtered = [result for result in results if result['lang'] == lang]
    if filtered:
        print("\nنتائج البحث:")
        for result in filtered:
            print(f"\nسؤال: {result['question']}\nإجابة: {result['answer']}\nتشابه: {result['similarity']:.2f}")
    else:
        fallback = results[0]
        msg = "لا توجد نتائج باللغة المختارة، إليك أفضل إجابة بلغة أخرى:" if lang == "ar" else "No results in the selected language; here is the best alternative answer:"
        print(f"\n{msg}\nسؤال: {fallback['question']}\nإجابة: {fallback['answer']}\nتشابه: {fallback['similarity']:.2f}")
```
