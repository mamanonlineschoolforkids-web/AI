## Setup

```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from model.faq_bot import FAQBot
```

## Load model

```python
bot = FAQBot(data_path="../model/data/faq_data.json")

print("✅ Bot loaded successfully")
```

## Test AR

```python
query = "ما هو مأمن؟"

results = bot.find_similar(query, top_k=3)

for r in results:
    print("Question:", r["question"])
    print("Answer:", r["answer"])
    print("Similarity:", round(r["similarity"], 2))
    print("-" * 40)
```

## Test EN

```python
query = "What is Ma'man?"

results = bot.find_similar(query, top_k=3)

for r in results:
    print("Question:", r["question"])
    print("Answer:", r["answer"])
    print("Lang:", r["lang"])
    print("Similarity:", round(r["similarity"], 2))
    print("-" * 40)
```

## Language detection

```python
text = "ما هو مأمن؟"

print("Detected language:", bot.detect_language(text))
```

## Chat demo

```python
while True:
    user_input = input("\n💬 اكتب سؤالك (exit للخروج): ")

    if user_input.lower() == "exit":
        break

    lang = bot.detect_language(user_input)
    print("🌍 Language:", lang)

    results = bot.find_similar(user_input, top_k=3)

    if not results:
        print("❌ No answer found")
        continue

    best = results[0]

    print("\n🤖 Answer:")
    print(best["answer"])
    print(f"⭐ Similarity: {best['similarity']:.2f}")
```
