```python
from preprocess import load_and_preprocess
from feature_engineering import create_features
from tfidf_recommender import build_tfidf, recommend_tfidf
from rule_based import rule_based_recommend
from user_profile import recommend_user_profile
from hybrid_recommender import hybrid_recommend
```

```python

# تحميل البيانات وتنظيفها
users, interactions, content = load_and_preprocess('users.csv', 'interactions.csv', 'content.csv')

# عمل الـ features
content, user_profiles = create_features(users, interactions, content)

# بناء TF-IDF
vectorizer, tfidf_matrix = build_tfidf(content)
```

```python
print("TF-IDF Recommender:")
recommend_tfidf(content_id=1, content=content, tfidf_matrix=tfidf_matrix, top_n=5)
```

```python

print("\nRule-Based Recommender:")
rule_based_recommend(age=10, content_id=1, content=content, tfidf_matrix=tfidf_matrix, top_n=5)
```

```python
print("\nUser Profile Recommender:")
recommend_user_profile(user_id=5, user_profiles=user_profiles, content=content, tfidf_matrix=tfidf_matrix, top_n=5)
```

```python
print("\nHybrid Recommender:")
hybrid_recommend(user_id=5, content_id=1, user_profiles=user_profiles, content=content, tfidf_matrix=tfidf_matrix, top_n=5)
```
